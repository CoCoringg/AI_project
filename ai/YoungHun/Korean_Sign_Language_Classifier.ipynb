# 2. 단어 목록의 단어별로 엑셀 파일을 분류

````python
# 영상 목록을 읽음
annotation = pd.read_csv('KETI-2017-SL-Annotation-v2_1.csv', encoding='utf-8')
annotation2 = pd.read_csv("KETI-2018-SL-Annotation-v1.csv", encoding="utf-8")
````

````python
# 목록을 번호, 파일명, 한국어만 남기고 다 없애버린다.
annotation = annotation[["번호", "파일명", "한국어"]]

# 목록을 번호, 파일명, 한국어만 남기고 다 없앤다.
annotation2 = annotation2[["번호", "파일명", "한국어"]]
````

````python
# 단어 목록에 있는 단어들을 추출해서 저장한다.
file_list = {key:None for key in words.keys()}

file_list2 = {key:None for key in words.keys()}

for word in words.keys():
    file_list[word] = annotation[annotation["한국어"] == word]
    file_list2[word] = annotation2[annotation2["한국어"] == word]
````

# 3. 동영상을 찾아서 이동

````python
# 폴더 생성
DATA_PATH = os.path.join("video_data")
for word in words.keys():
    try:
        os.makedirs(os.path.join(DATA_PATH, words[word]))
    except:
        pass
````

### 9000번 이전 영상 폴더 분리

````python
# 동영상 파일을 찾아서 이동
for word in words.keys():
    for i in range(len(file_list[word])):
        video_num, video_name, _ = file_list[word].iloc[i]
        
        video_num = int(video_num)
        video_name = video_name[:-3] + "avi"
        video_path_name = f"{i}.avi"
        # 동영상이 어떤 폴더에 들어있는지 확인
        folder_name = None
        if video_num <= 3000:
            folder_name = "0001~3000"
        elif video_num <= 6000:
            folder_name = "3001~6000"
        elif video_num <= 8380:
            folder_name = "6001~8380"
        elif video_num <= 9000:
            folder_name = "8381~9000"
        elif video_num <= 10480:
            folder_name = "9601~10480"
        
        if folder_name is None:
            print(f"folder not found for {video_num}")
            continue
        
        # 동영상을 옮김
        video_path_cur = os.path.join(folder_name, video_name)
        video_path_mov = os.path.join(DATA_PATH, words[word], video_path_name)
        try:
            shutil.move(video_path_cur, video_path_mov)
            print(f"{video_path_name} moved")
        except:
            print(f"{video_name} failed")
        
````
